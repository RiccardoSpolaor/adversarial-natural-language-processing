```python
import pickle,os
import matplotlib.pyplot as plt
import numpy as np
```

```python
short_reviews_negative_experiments_dictionary = {}
short_reviews_positive_experiments_dictionary = {}

long_reviews_negative_experiments_dictionary = {}
long_reviews_positive_experiments_dictionary = {}
```

```python
base_short_reviews_dir='./small_adversarial_reviews/'
base_long_reviews_dir='./big_adversarial_reviews/'
neg_dir = 'neg/'
pos_dir = 'pos/'
```

```python
class Hyperparameter:
    MAX_WORDS = 'maxwords'
    TOP_NEIGHBOURS = 'topneighbours'
    TOP_LM = 'toplm'
    MAX_CHILDREN = 'maxchildren'
    MAX_POP_MEMBERS = 'maxpopmembers'
    
class StatKind:
    EVALUATION = 'Attack Success'
    TIME = 'Time'
    GENERATIONS = 'Generations'
    
class ReviewKind:
    SHORT = 'Short'
    LONG = 'Long'
```

```python
def open_pickle_file(name, maxwords, topneighbours, toplm, maxchildren, maxpopmembers):
    
    with open(os.path.join(base_short_reviews_dir + neg_dir + 'maxwords{}_topneighbours{}_toplm{}_maxchildren{}_maxpopmembers{}.pickle'.format(
        maxwords, 
        topneighbours, 
        toplm,
        maxchildren,
        maxpopmembers
    )), 'rb') as f:
        short_reviews_negative_experiments_dictionary[name] = pickle.load(f)
    f.close()

        
    with open(os.path.join(base_long_reviews_dir + neg_dir + 'maxwords{}_topneighbours{}_toplm{}_maxchildren{}_maxpopmembers{}.pickle'.format(
        maxwords, 
        topneighbours, 
        toplm,
        maxchildren,
        maxpopmembers
    )), 'rb') as f:
        long_reviews_negative_experiments_dictionary[name] = pickle.load(f)
    f.close()
    
    with open(os.path.join(base_short_reviews_dir + pos_dir + 'maxwords{}_topneighbours{}_toplm{}_maxchildren{}_maxpopmembers{}.pickle'.format(
        maxwords, 
        topneighbours, 
        toplm,
        maxchildren,
        maxpopmembers
    )), 'rb') as f:
        short_reviews_positive_experiments_dictionary[name] = pickle.load(f)
    f.close()
    
    with open(os.path.join(base_long_reviews_dir + pos_dir + 'maxwords{}_topneighbours{}_toplm{}_maxchildren{}_maxpopmembers{}.pickle'.format(
        maxwords, 
        topneighbours, 
        toplm,
        maxchildren,
        maxpopmembers
    )), 'rb') as f:
        long_reviews_positive_experiments_dictionary[name] = pickle.load(f)
    f.close()
```

```python
def get_plot_values(hyperparameter):
    if hyperparameter == Hyperparameter.MAX_WORDS: 
        return ['10', '15']
    elif hyperparameter == Hyperparameter.TOP_NEIGHBOURS: 
        return ['15', '20'] 
    elif hyperparameter == Hyperparameter.TOP_LM: 
        return ['6', '8']
    elif hyperparameter == Hyperparameter.MAX_CHILDREN: 
        return ['4', '6']
    elif hyperparameter == Hyperparameter.MAX_POP_MEMBERS: 
        return ['10', '15']
    else:
        return None

def set_plot_title(stat_kind, hyperparameter):
    if stat_kind == StatKind.TIME or stat_kind == StatKind.GENERATIONS:
        return 'Mean {} Results for {}'.format(stat_kind, hyperparameter)
    elif stat_kind == StatKind.EVALUATION:
        return '{} for {}'.format(stat_kind, hyperparameter)

def get_plot_data (hyperparameter, stat_kind, values, review_kind):
    
    if review_kind == ReviewKind.SHORT:
        positive_experiments_dictionary = short_reviews_positive_experiments_dictionary
        negative_experiments_dictionary = short_reviews_negative_experiments_dictionary
    elif review_kind == ReviewKind.LONG:
        positive_experiments_dictionary = long_reviews_positive_experiments_dictionary
        negative_experiments_dictionary = long_reviews_negative_experiments_dictionary
    
    if stat_kind == StatKind.EVALUATION:
        res = 'evaluation'
    elif stat_kind == StatKind.TIME:
        res = 'time_list'
    elif stat_kind == StatKind.GENERATIONS:
        res = 'generations_list'
        
    positive_data = []
    negative_data = []
        
    if stat_kind == StatKind.GENERATIONS or stat_kind == StatKind.TIME:
        positive_data += [sum(positive_experiments_dictionary['base'][res]) / len(positive_experiments_dictionary['base'][res])]
        for value in values:
            positive_data += [
                sum(positive_experiments_dictionary[hyperparameter + value][res]) / len(positive_experiments_dictionary[hyperparameter + value][res])
            ]
        
        negative_data += [sum(negative_experiments_dictionary['base'][res]) / len(negative_experiments_dictionary['base'][res])]
        for value in values:
            negative_data += [
                sum(negative_experiments_dictionary[hyperparameter + value][res]) / len(negative_experiments_dictionary[hyperparameter + value][res])
            ]
    elif stat_kind == StatKind.EVALUATION:
        positive_data += [(1 - positive_experiments_dictionary['base'][res])*100]
        for value in values:
            positive_data += [(1 - positive_experiments_dictionary[hyperparameter + value][res]) * 100]
        
        negative_data += [(1 - negative_experiments_dictionary['base'][res])*100]
        for value in values:
            negative_data += [(1 - negative_experiments_dictionary[hyperparameter + value][res]) * 100]
            
    return positive_data, negative_data

def value_label(ax, rectangles):
    offset = {'C' : 0, 'R' : 1, 'L' : -1}
    for rect in rectangles:
        height = rect.get_height()
        ax.annotate(
            '{:.2f}'.format(height),
            xy = (rect.get_x() + rect.get_width() / 2, height),
            xytext = (0, 3),
            textcoords = 'offset points',
            ha = 'center', va = 'bottom'
        )
```

```python
def plot_parameters(hyperparameter, stat_kind, review_kind):
    
    base_values = {
        Hyperparameter.MAX_WORDS: 5,
        Hyperparameter.TOP_NEIGHBOURS: 10,
        Hyperparameter.TOP_LM: 4,
        Hyperparameter.MAX_CHILDREN: 2,
        Hyperparameter.MAX_POP_MEMBERS: 5
    }
    
    values = get_plot_values(hyperparameter)
    
    index = np.arange(3)
    positive_data, negative_data = get_plot_data (hyperparameter, stat_kind, values, review_kind)
    
    fig, ax = plt.subplots(figsize=(15,10))
    
    bar_width = 0.2
    opacity = 0.8
    
    rectangles_1 = plt.bar(index, positive_data, bar_width, color = 'g', label = 'Positive Reviews', alpha = opacity)
    rectangles_2 = plt.bar(index + bar_width, negative_data, bar_width, color = (0.95,0,0), label = 'Negative Reviews', alpha = opacity)
    
    plt.xlabel(hyperparameter)
    
    if stat_kind == StatKind.TIME:
        result_string = 'Mean ' + stat_kind + ' (seconds)'
    elif stat_kind == StatKind.GENERATIONS:
        result_string = 'Mean ' + stat_kind
    elif stat_kind == StatKind.EVALUATION:
        result_string = stat_kind + ' (%)'
    
    plt.ylabel(result_string)
    
    plt.title(set_plot_title(stat_kind, hyperparameter))
    
    plt.xticks(index + bar_width/2, ([base_values[hyperparameter]] + values))
    
    value_label(ax, rectangles_1)
    value_label(ax, rectangles_2)
    
    plt.legend()
    
    plt.show()
```

## Hyperparameters Evaluation

The testing procedure consists in checking how the **mean generations** and **time** to generate an *Adversarial Example* vary when the hyperparameters value changes. The **attack success ratio** with different hyperparameters combinations is also tested.

The following hyperparameters were considered for the test: `max_words`; `top_neighbours`; `top_lm`; `max_children`; `max_pop_members`.

Their default values are the following ones:
* `max_words`: 5,
* `top_neighbours`: 10,
* `top_lm`: 4,
* `max_children`: 2,
* `max_pop_members`: 5.

For each hyperparameter two more possible values have been considered and tested while maintaining the other hyperparameters at their default values.

## Evaluation for Reviews of Different Length

It is then calculated the **average generations** and **time** needed to generate an Adversarial Example and the **mean attack success ratio** considering all the hyperparameters combinations for both *short reviews* (maximum 500 characters) and *long reviews* (more than 300 characters, but less than 2,000).

```python
def set_plot_title(stat_kind):
    common_string = 'Review Kind'
    if stat_kind == StatKind.TIME or stat_kind == StatKind.GENERATIONS:
        return 'Mean {} Results for {}'.format(stat_kind, common_string)
    elif stat_kind == StatKind.EVALUATION:
        return '{} for {}'.format(stat_kind, common_string)


def get_stat_results_on_review_kind (res, stat_kind, review_kind, sentiment_kind):
    
    if review_kind == ReviewKind.SHORT:
        if sentiment_kind == 0:
            experiments_dictionary = short_reviews_negative_experiments_dictionary
        else:
            experiments_dictionary = short_reviews_positive_experiments_dictionary
        
    elif review_kind == ReviewKind.LONG:
        if sentiment_kind == 0:
            experiments_dictionary = long_reviews_negative_experiments_dictionary
        else:
            experiments_dictionary = long_reviews_positive_experiments_dictionary
    
    if stat_kind == StatKind.GENERATIONS or stat_kind == StatKind.TIME:
        
        return sum(
                [sum(experiments_dictionary[key][res]) 
                 for key in experiments_dictionary.keys()
                ]
            )/sum(
                [len(experiments_dictionary[key][res]) 
                 for key in experiments_dictionary.keys()
                ]
            )
    elif stat_kind == StatKind.EVALUATION:
        
        return sum(
                [(1 - experiments_dictionary[key][res])*100 
                 for key in experiments_dictionary.keys()
                ])/len(experiments_dictionary)
    
        '''
        
        negative_data += [(1 - negative_experiments_dictionary['base'][res])*100]
        for value in values:
            negative_data += [(1 - negative_experiments_dictionary[hyperparameter + value][res]) * 100]
    '''
def get_plot_data (stat_kind):
    if stat_kind == StatKind.EVALUATION:
        res = 'evaluation'
    elif stat_kind == StatKind.TIME:
        res = 'time_list'
    elif stat_kind == StatKind.GENERATIONS:
        res = 'generations_list'
        
    positive_data = [get_stat_results_on_review_kind (res, stat_kind, ReviewKind.SHORT, 1),
                    get_stat_results_on_review_kind (res, stat_kind, ReviewKind.LONG, 1)]
    negative_data = [get_stat_results_on_review_kind (res, stat_kind, ReviewKind.SHORT, 0),
                    get_stat_results_on_review_kind (res, stat_kind, ReviewKind.LONG, 0)]
        
    return positive_data, negative_data

def value_label(ax, rectangles):
    offset = {'C' : 0, 'R' : 1, 'L' : -1}
    for rect in rectangles:
        height = rect.get_height()
        ax.annotate(
            '{:.2f}'.format(height),
            xy = (rect.get_x() + rect.get_width() / 2, height),
            xytext = (0, 3),
            textcoords = 'offset points',
            ha = 'center', va = 'bottom'
        )
```

```python
def plot_short_long_reviews_stats(stat_kind):
    
    index = np.arange(2)
    positive_data, negative_data = get_plot_data (stat_kind)
    
    fig, ax = plt.subplots(figsize=(15,10))
    
    bar_width = 0.2
    opacity = 0.8
    
    rectangles_1 = plt.bar(index, positive_data, bar_width, color = 'g', label = 'Positive Reviews', alpha = opacity)
    rectangles_2 = plt.bar(index + bar_width, negative_data, bar_width, color = (0.95,0,0), label = 'Negative Reviews', alpha = opacity)
    
    plt.xlabel('Review Kind')
    
    if stat_kind == StatKind.TIME:
        result_string = 'Mean ' + stat_kind + ' (seconds)'
    elif stat_kind == StatKind.GENERATIONS:
        result_string = 'Mean ' + stat_kind
    elif stat_kind == StatKind.EVALUATION:
        result_string = stat_kind + ' (%)'
    
    plt.ylabel(result_string)
    
    plt.title(set_plot_title(stat_kind))
    
    plt.xticks(index + bar_width/2, (['reviews shorter than 500 characters', 'reviews longer than 300 characters, but shorter than 2,000']))
    
    value_label(ax, rectangles_1)
    value_label(ax, rectangles_2)
    
    plt.legend()
    
    plt.show()
```
